**INICIO DE CARGA DE ARCHIVO CSV**

```python
#Carga desde un archivo .csv sin indice
df =pd.read_csv('Francia.csv')
df
#Inmuebles['Columna']=Inmuebles['Columna'].str.replace('%', '')
```

**FILTRO PARA CINCO "HOST NAME" DIFERENTES**

```python
#Agarrar un nombre de la columna de host_name
#filtro1=df.iloc[[5,8,12,34,50],3]   
#filtro1
filtro1=df[(df["host_name"]=="Isabelle")| (df["host_name"]=="Sandrine")| (df["host_name"]=="Polali")| (df["host_name"]=="Pablo")| (df["host_name"]=="Adrian")]
filtro1
```

```python
#Convertir archivo filtrado a CSV
filtro1.to_csv('Filtro_Host_Name.csv')
```

**FILTRO PARA LOS HOST QUE SE HAYAN UNIDO A AIRBNB "DESPUÉS DEL AÑO 2020"**

```python
filtro2=df[(df["host_since"]>="2020-01-01")]
filtro2
```

```python
#Convertir archivo filtrado a CSV
filtro2.to_csv('Filtro_Host_Since.csv')
```

**FILTRO PARA LOS HOST QUE RESPONDEN "A MÁS TARDAR EN 1 DÍA"**

```python
filtro3=df[(df["host_response_time"]=="within a day")| (df["host_response_time"]=="within an hour")|(df["host_response_time"]=="within a few hours")]
filtro3  
```

```python
#Convertir archivo filtrado a CSV
filtro3.to_csv('Filtro_Host_Response.csv')
```
