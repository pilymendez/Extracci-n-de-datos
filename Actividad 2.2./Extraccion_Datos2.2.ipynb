Los registros de los host cuyo "host_response_rate" sea mayor a 95% y que no sea "superhost"

```python
filtro1=df[(df["host_response_rate"]>=0.95) & (df["host_is_superhost"]=="f")]
filtro1
```

```python
#Convertir archivo filtrado a CSV
filtro1.to_csv('01_host_response_rate.csv')
```

Los registros cuyas propiedades no tengan "1 baño"

```python
filtro2=df[~(df["bathrooms"]==1)] #Niega todo
filtro2
```

```python
#Convertir archivo filtrado a CSV
filtro2.to_csv('02_no_bathroom.csv')
```

Los registros que tengan un precio entre $100 y $200 

```python
filtro3=df[(df["price"]<=200) & (df["price"]>=100)]
filtro3
```

```python
#Convertir archivo filtrado a CSV
filtro3.to_csv('03_precio_100_200.csv')
```

Los registros que sean "instant_bookable" y que hallan verificado su perfil "host_identify_verified"

```python
filtro4=df[(df["instant_bookable"]=="t") & (df["host_identity_verified"]=="t")]
filtro4
```

```python
#Convertir archivo filtrado a CSV
filtro4.to_csv('04_instant_bookable.csv')
```

Los registros de los tipos de cuarto: "hotel room" o "private room"

```python
filtro5=df[(df["room_type"]=="Hotel room") | (df["room_type"]=="Private room")]
filtro5
```

```python
#Convertir archivo filtrado a CSV
filtro5.to_csv('05_tipos_cuarto.csv')
```

Los que tengan mas de 4 accommodates

```python
filtro6=df[(df["accommodates"]>4)]
filtro6
```

```python
#Convertir archivo filtrado a CSV
filtro6.to_csv('06_accomodates.csv')
```
